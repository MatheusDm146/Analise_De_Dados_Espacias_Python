```python
#importando bibliotecas necessarias!
import pandas as pd
import geopandas as gpd
import folium
```

```python
#lendo o arquvivo CSV
#contendo a distribuição global de vulcoes em 2010
df = pd.read_csv(r"C:\Users\Usuario\OneDrive\Desktop\volcano_data_2010.csv")

# Deixando somentes as colunas necessarias
df = df[["Year","Name","Country","Latitude","Longitude","Type"]]
df
```

```python
#Criando geometrias de pontos (longitude e latitude) para cada vulcão
geometrias = gpd.points_from_xy(df["Longitude"], df["Latitude"])

#usando um GeoDataFrama a partir de um Data Frame e dos dados de geometria
geo_df = gpd.GeoDataFrame(df, geometry=geometrias)
geo_df
```

```python
#criando um mapa interativo com Folium

import folium.plugins


mapa = folium.Map(Location=[15, 30], tiles="Cartodb dark_matter", zoom_start=2)
mapa

# Criando uma lista com os pontos de erupção vulcanicas a partir do GeoDataframe
heatmap_dados = [[point.xy[1][0], point.xy[0][0]] for point in geo_df["geometry"]]

# Criando heatmap para a visualizar a densidade de vulcoes
folium.plugins.HeatMap(heatmap_dados).add_to(mapa)


#vizualizando mapa
mapa
```
